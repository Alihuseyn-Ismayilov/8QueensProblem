```python
import numpy as np
import random
t = 0 #i tried to do everything step by step according to our lesson
n = int(input("Enter Board Size"))#user give board size
population_size = int(input("Enter Population size, I have tried many numbers and 50 is fastest"))#user can type population size
fitness_values = []#stoeing fitness values in this list
result_list = []#storing list of solution in this list

population =  list(list(random.randint(1,n) for i in range(n))for i in range(population_size)) #create random population(size=n)

def check_attack(given_list):#it is for calculating nonatacking queens for given list of population
    count = 0
    for i in range(1,len(given_list)+1):#here simply i check 3 conditions if (rows, cols or (row1-row2) = (col1- col2) are same
        for z in range(i+1,len(given_list)+1):#i use abs for checking whole diagonal(row - col)
            if( (i == z) or  ( given_list[i-1] == given_list[z-1] ) or (abs( (i - z) ) == abs( given_list[i-1] - given_list[z-1] ) )  ):
                pass#if selected queen attack in these positions to another selected queen(repeated for all qeens) then pass
            else:
                count+=1#if doenst attack then increase count
    return count

def fitness_function(population):
    fitness_list = []
    for i in range(len(population)):
        fitness_list.append(check_attack(population[i]))
    return fitness_list#i call check_attack function for every index of population

def select_parents(fitness_values):#i take 2 different list from population for next generation in every  iteration(half of population size)
    global population
    population1 = []
    parents = [0,0]
    #temp_list_percentage = []
    #temp_list_percentage = percentage_fitness(fitness_values)
    for i in range(population_size//2):
        while parents[0] == parents[1]:#if parents are same then iterate untill parents  will be diferente
            parents = random.choices(population,k = 2)#take 2 list from population
        population1.append(parents[0])#including it to new created list(population1)
        population1.append(parents[1])#including it to new created list(population1)
        parents = [0,0]#i made it 0 because of condition of while
    population = population1#now population1 is list of parents and we make  population equal to population1
    return population#here return 0 can be written , we dont need what return give us

def recombine():#i combine parents with crossover , for generating new population
    global population
    temp = []
    crossover = 0
    z=0
    for i in range(0,len(population),2):
        z = i+1
        crossover = random.randint(1,n-1)#choose number between 1, n-1 because we need at least 1 number  that doesnt change
        temp = population[i][::]#using temp for changing population items according to crossover point
        population[i][crossover:] = population[z][crossover:]
        population[z][crossover:] = temp[crossover:]
        temp = [0]
    return 0

def random_mutation():#rate of mutation is 3% and it happens randomly
    global population
    for i in range(len(population)):
        if(np.random.random() < 0.03):#0.1
            population[i][random.randint(0,n-1)] = random.randint(1,n)
    return population  







fitness_values = fitness_function(population)

print("waiting for solution...." )#comment for user
while( (n*(n-1))/2 != max(fitness_values)):#main loop , it continues untill finding best solution
    t = t+1
    select_parents(fitness_values)#call parents function for choosing parents
    recombine()#call it for combine parents
    random_mutation()#call it for mutate randomly choosen number of random list
    fitness_values = fitness_function(population)
print("solution found!!")
result_list = population[fitness_values.index(max(fitness_values))]
print("In" , t , "th Generation fitness value is " , max(fitness_values))
print(result_list)
```
